#### data cleaning with regex

```python
dirtyMat = [(0,'monday','<efs> sfsdc <sefsdvdv> erfsdvdrer.'),
           (1,'tuesday','sdkjnvkj 234 5t  <fdbdf> dvcvdfv$%^ <df> wefdfvd')]
```

```python
dirtyDF = pd.DataFrame(dirtyMat)
```

```python
dirtyDF
```

```python
import re
re.sub('<\S+>','','<efs> sfsdc <sefsdvdv> erfsdvdrer.')
```

```python
dirtyDF[2] = dirtyDF[2].apply(lambda x : re.sub('<\S+>','',x))
```
